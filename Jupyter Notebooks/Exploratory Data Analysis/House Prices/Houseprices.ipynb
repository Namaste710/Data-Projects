# Variables and Imports

```python
TRAIN_PATH = 'data/train.csv'
TEST_PATH = 'data/test.csv'
import pandas as pd
train_df = pd.read_csv(TRAIN_PATH)
test_df = pd.read_csv(TEST_PATH)
train_df.head()
```

```python
limitNA = len(train_df) * .80
train_df = train_df.dropna(thresh=limitNA, axis=1)
test_df = test_df.dropna(thresh=limitNA, axis=1)
train_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
def add_cat_cols(input_df):
    non_numeric_columns = input_df.select_dtypes(exclude=['int64', 'float64']).columns
    numeric_columns = input_df.select_dtypes(include=['int64', 'float64']).columns
    le = LabelEncoder()
    non_numeric_columns = input_df.select_dtypes(exclude=['int64', 'float64']).columns
    numeric_columns = input_df.select_dtypes(include=['int64', 'float64']).columns
    cat_df = input_df[non_numeric_columns].fillna('Empty') # muss auf jeden Fall woanders hin
    for column in non_numeric_columns:
        input_df[column + '_Cat'] = le.fit_transform(cat_df[column])
    transformed_non_numeric_columns = [col for col in input_df.columns if '_Cat' in col]
    selected_cols = numeric_columns.to_list()
    selected_cols.extend(transformed_non_numeric_columns)
    input_df.drop(columns=[col for col in input_df if col not in selected_cols], inplace=True)
    return input_df
```

```python
train_df = add_cat_cols(train_df)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
X_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, -1:].values
```

```python
X_train = imp_mean.fit_transform(X_train)
y_train = imp_mean.fit_transform(y_train)
```

```python
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (20, 12)})
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

importance = lr_model.coef_

print(importance)
```

```python
feature_importances = pd.DataFrame({'Feature': train_df.columns.drop('Id').to_list(), 'Importance value': importance.flatten()})
```

```python
feature_importances.sort_values('Importance value', ascending=False, inplace=True)
```

```python
top15 = feature_importances.nlargest(15, 'Importance value')
top12 = feature_importances.nlargest(12, 'Importance value')
top9 = feature_importances.nlargest(9, 'Importance value')
```

```python
sns.barplot(top15['Feature'], top15['Importance value'])
```

```python
sns.barplot(top12['Feature'], top12['Importance value'])
```

```python
sns.barplot(top9['Feature'], top9['Importance value'])
```

```python
picked_cols = list(top9['Feature'])

picked_train_df = train_df[picked_cols]

X_train_top9 = picked_train_df.iloc[:, :-1].values
y_train_top9 = picked_train_df.iloc[:, -1:].values

X_train_top9 = imp_mean.fit_transform(X_train_top9)
y_train_top9 = imp_mean.fit_transform(y_train_top9)

lr_model_2 = LinearRegression()

lr_model_2.fit(X_train_top9, y_train_top9)
```

```python
test_df = add_cat_cols(test_df)
picked_test_df = test_df[picked_cols]

X_test_top9 = picked_test_df.iloc[:, :-1].values
y_test_top9 = picked_test_df.iloc[:, -1:].values

X_test_top9 = imp_mean.fit_transform(X_test_top9)
y_test_top9 = imp_mean.fit_transform(y_test_top9)

y_pred_top9 = lr_model_2.predict(X_test_top9)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test_top9, y_pred_top9)
```
